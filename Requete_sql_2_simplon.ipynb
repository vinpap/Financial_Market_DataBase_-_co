```python
import pandas as pd
import sqlite3
import yfinance as yf
```

```python
# create a sql connection and create the database stock
c = sqlite3.connect("stock.db")
```

On crée une table pour les prix 

```python
# create price table with the primary tuple keys ( date, ticker) and the columns, date, ticker, price
query1 = """CREATE TABLE IF NOT EXISTS price (
        date DATE NOT NULL,
        ticker varchar(255) NOT NULL,
        price FLOAT,
        PRIMARY KEY(date, ticker)
        )"""
c.execute(query1.replace('\n',' '))
```

```python
# create volume table with the primary tuple keys ( date, ticker) and the columns, date, ticker, volume 
query2 = """CREATE TABLE IF NOT EXISTS volume (
Date TEXT NOT NULL,
ticker TEXT NOT NULL,
volume REAL,
PRIMARY KEY(Date, ticker)
)"""
c.execute(query2.replace("\n", " "))
```

```python
# backtest inputs
bt_inputs = {'tickers': ['BA', 'UNH', 'MCD', 'HD'], #BA: Boeing, UNH: UnitedHealth, MCD: Mac Donalds, HD : HomeDepot
'start_date': '2019-01-01',
'end_date': '2021-06-01'}
```

```python
def download(bt_inputs, proxy = None):
    data = yf.download(tickers= bt_inputs['tickers'],
                       start = bt_inputs['start_date'],   
                       end = bt_inputs['end_date'],
                       interval = '1d',
                       prepost = True,
                       threads = True,
                       proxy = proxy)
    return data
```

```python
test = download(bt_inputs)
```

```python
# inputs
select_tickers = bt_inputs['tickers']
start_date = bt_inputs['start_date']
end_date = bt_inputs['end_date']
# construct query: select the prices withe tickers in select_tickers 
#between start_date and end_date
query = f"""SELECT ticker, price FROM price WHERE ticker in ({select_tickers}) AND {start_date} < Date < {end_date}"""
print("Query: ", query.replace('\n',' ').replace("[", "").replace("]", ""))
cur = c.execute(query.replace('\n',' ').replace("[", "").replace("]", ""))
```

```python
#display the request
rows = cur.fetchall()
print(f"{len(rows)} rows found")
print(rows)
```

```python
#store the results of the request in a  dataframe result
result = pd.DataFrame(rows, columns=["ticker", "price"])
result
```
